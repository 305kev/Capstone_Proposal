```python
import seaborn as sns

data = {'time': {'Up to 15%': 47.3, '15% - 25 %': 24.3, '25% - 50%': 12.3,
                        '50% - 75%': 7.0, 'More than 75%': 3.3, 'None': 5.7}}
```

```python
plt.figure(figsize=(10,10))
plt.title('Average Time Spent on Research', size=30)

df_data = pd.DataFrame(data).sort_values('time', ascending=False)
b= sns.barplot(x='time', y=df_data.index, data=df_data, palette=None)
b.set_xlabel("Percent of surveyed lawyers",fontsize=25)
b.set_ylabel("Percent of time",fontsize=25, rotation=0,  y=.5)
plt.yticks(rotation=0, size=20) 
plt.xticks(rotation=0, size=20) 

```

```python
import matplotlib.pyplot as plt
 

# data = {'time': {'Up to 15%': 47.3, '15% - 25 %': 24.3, '25% - 50%': 12.3,
#                         '50% - 75%': 7.0, 'More than 75%': 3.3, 'None': 5.7}}
# Data to plot
labels = 'Up to 15%', '15% - 25 %', '25% - 50%', '50% - 75%', 'More than 75%', 'None'
sizes = [47.3, 24.3, 12.3, 7.0, 3.3, 5.7]

print(len(sizes), len(labels))

colors = ['gold', 'gold', 'gold', 'gold', 'gold', 'violet']
explode = (0, 0, 0, 0, 0, 0.5)  # explode 1st slice
 
# Plot
plt.rcParams['font.size'] = 20.0
plt.figure(figsize=(10,10))
plt.title("Average Time Spent on Legal Research*", size=30)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

 
plt.axis('equal')
plt.show()
```
